## Data Cleaning

### Filling Missing Value


```python

import pandas as pd

# Reload after reset
crime = crime_incident.copy()

# Generate missing value report
missing_report = crime.isna().sum().to_frame("missing_count")
missing_report["missing_pct"] = (missing_report["missing_count"] / len(crime) * 100).round(2)
missing_report_sorted = missing_report.sort_values("missing_pct", ascending=False)

import caas_jupyter_tools
caas_jupyter_tools.display_dataframe_to_user("Crime Incident Missing Value Report", missing_report_sorted)

```
